```python
import pandas as pd
import numpy as np
import skfuzzy as fuzz
from sklearn.preprocessing import StandardScaler, LabelEncoder
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("respiratory_clean.csv")
```

```python
data.head(2)
```

```python
# Inspect the gender column
print("Unique values in 'gender' column:", data['sex'].unique())
```

```python
# Apply label encoding
label_encoder = LabelEncoder()
data['sex_encoded'] = label_encoder.fit_transform(data['sex'])
```

```python
# Display the mapping of labels
label_mapping = dict(zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_)))
print("Label Encoding Mapping:", label_mapping)
```

```python
# Drop the original gender column (optional)
data.drop(columns=['sex'], inplace=True)

# Check the updated dataframe
data.head(2)
```

```python
data.columns
```

```python
features = ['age', 'duration', 'cough', 'night_fever', 'catarrh', 'cold',
       'tenderness', 'headache', 'black_stool',
       'jerking_movement_of_the_hands', 'unable_to_sit_unsupported',
       'throat_fever', 'fast_breathing', 'body_rashes', 'cough_sills',
       'frequent_stooling', 'chest_pain', 'frequent_vomiting',
       'sleep_disturbance_at_night', 'nasal_stuffiness', 'excessive_sweating',
       'chronic_cough', 'profuse_nasal_discharge', 'difficulty_in_breathing',
       'itching_ear', 'frequent_urination',
       'swelling_of_pubic_area_right_sided', 'generalized_body_rashes',
       'painful_swallowing', 'passage_of_freight', 'poor_vision',
       'frequent_micturition', 'watery_of_the_tongue', 'poor_appetite',
       'noisy_breathing', 'pepsis_swells_up_since_birth', 'body_weakness',
       'flame_pain', 'swollen_leg', 'abdominal_pain', 'excessive_salivation',
       'crawling_sensation', 'weight_loss', 'g.b_pain', 'stomach_ache',
       'tysing_of_the_right_ear', 'refusal_of_food', 'waist_pain', 'snoring',
       'knee_pain', 'dizziness', 'scrotal_swelling', 'vaginal_discharge',
       'vaginal_itching', 'stooling', 'running_nose', 'club_pain',
       'bleeding_from_the_nose', 'unable_to_speak', 'mouth_breathing',
       'sleep_apneas', 'sex_encoded']  # Adjust based on actual feature names
data_features = data[features]
```

```python
scaler = StandardScaler()
data_normalized = scaler.fit_transform(data_features)
```

```python
# Set parameters for Fuzzy C-Means
n_clusters = 2  # Define the number of clusters
fuzziness = 2.0  # Typically 2.0 is used
max_iter = 150  # Maximum number of iterations
error = 1e-5  # Error tolerance

# Apply Fuzzy C-Means clustering
cntr, u, u0, d, jm, p, fpc = fuzz.cluster.cmeans(
    data_normalized.T, n_clusters, fuzziness, error=error, maxiter=max_iter, init=None
)

# Assign each data point to a cluster with maximum membership
cluster_labels = np.argmax(u, axis=0)
```

```python
print('Cluster Centers:')
print(cntr)
```

```python
# Add the cluster labels back to the original dataframe
data['Cluster'] = cluster_labels
```

```python
# Save the clustered data
data.to_csv('respiratory_clustered.csv', index=False)
```

```python
# Visualize the clusters (optional - replace 'Feature1', 'Feature2' with actual features for visualization)
plt.figure(figsize=(10, 7))
for i in range(n_clusters):
    plt.scatter(data[data['Cluster'] == i][features[0]], 
                data[data['Cluster'] == i][features[1]], 
                label=f'Cluster {i}')
plt.xlabel(features[0])
plt.ylabel(features[1])
plt.title('Fuzzy C-Means Clustering Results')
plt.legend()
plt.show()
```

```python
# Output the fuzzy partition coefficient
print(f'Fuzzy Partition Coefficient (FPC): {fpc}')

# Analyze cluster centers to understand characteristics of each cluster
cluster_centers = scaler.inverse_transform(cntr)
print('Cluster Centers:', cluster_centers)
```

```python
from sklearn.decomposition import PCA

# Reduce data to 2 components for visualization
pca = PCA(n_components=2)
data_pca = pca.fit_transform(data[features])

plt.figure(figsize=(10, 7))
for i in range(n_clusters):
    plt.scatter(data_pca[data['Cluster'] == i, 0],
                data_pca[data['Cluster'] == i, 1],
                label=f'Cluster {i}')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.title('Fuzzy C-Means Clustering Results (PCA Reduced)')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import silhouette_score

# Convert fuzzy memberships to hard labels for silhouette score
hard_labels = np.argmax(u, axis=0)
silhouette_avg = silhouette_score(data_normalized, hard_labels)
print(f'Silhouette Score: {silhouette_avg}')
```

```python
import seaborn as sns

# Plot histogram of membership values for each cluster
plt.figure(figsize=(12, 6))
for i in range(n_clusters):
    sns.histplot(u[i], kde=True, bins=20, label=f'Cluster {i} Membership')
plt.xlabel('Membership Degree')
plt.ylabel('Frequency')
plt.title('Distribution of Membership Degrees for Each Cluster')
plt.legend()
plt.show()
```
